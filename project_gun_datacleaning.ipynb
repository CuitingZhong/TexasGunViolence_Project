```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
import re
import warnings

warnings.filterwarnings('ignore')
%pylab inline
```

```python
df=pd.read_csv('texas.csv')
df.info()
```

```python
df['n_victim']=df['participant_type'].map(lambda x:len(re.findall('Victim',str(x))))
df['n_suspect']=df['participant_type'].map(lambda x:len(re.findall('Subject-Suspect',str(x))))
```

```python
victim_gender=list(df['participant_gender'].map(lambda x:re.findall('[a-zA-Z]+',str(x))))
for i in range(len(victim_gender)):
    j=df['n_victim'][i]
    if j==0:
        victim_gender[i]=NaN
    else:
        victim_gender[i]=victim_gender[i][0:j]
victim_gender=Series(victim_gender)
df['victim_gender']=victim_gender
```

```python
suspect_gender=list(df['participant_gender'].map(lambda x:re.findall('[a-zA-Z]+',str(x))))
for i in range(len(victim_gender)):
    j=df['n_suspect'][i]
    if j==0:
        suspect_gender[i]=NaN
    else:
        suspect_gender[i]=suspect_gender[i][-j:]
suspect_gender=Series(suspect_gender)
df['suspect_gender']=suspect_gender
```

```python
age_group=df['participant_age_group'].map(lambda x:re.findall('[a-zA-Z]+ \d+[-]*\d+[\+]*',str(x)))
len_each_age=age_group.map(lambda x:len(x))
idx=len_each_age[len_each_age==0].index.values
age_group[idx]=df.iloc[idx]['participant_age_group']
vic_age_g={}
suspect_age_g={}
for i in range(len(age_group)):
    j=int(df['n_suspect'][i])
    k=int(df['n_victim'][i])
    if i not in idx:
        if j==0:
            suspect_age_g[i]=NaN
        else:
            suspect_age_g[i]=age_group[i][-j:]
        if k==0:
            vic_age_g[i]=NaN
        else:
            vic_age_g[i]=age_group[i][0:k]
suspect_age_g=Series(suspect_age_g)
vic_age_g=Series(vic_age_g)
df['victim_age_group']=vic_age_g
df['suspect_age_group']=suspect_age_g
```

```python
df['n_victim_male']=df['victim_gender'].map(lambda x:len(re.findall('Male',str(x))))
df['n_victim_female']=df['victim_gender'].map(lambda x:len(re.findall('Female',str(x))))
df['n_suspect_male']=df['victim_gender'].map(lambda x:len(re.findall('Male',str(x))))
df['n_suspect_female']=df['victim_gender'].map(lambda x:len(re.findall('Female',str(x))))

df['n_suspect_Child']=df['suspect_age_group'].map(lambda x:len(re.findall('Child',str(x))))
df['n_suspect_Teen']=df['suspect_age_group'].map(lambda x:len(re.findall('Teen',str(x))))
df['n_suspect_Adult']=df['suspect_age_group'].map(lambda x:len(re.findall('Adult',str(x))))

df['n_victim_Child']=df['victim_age_group'].map(lambda x:len(re.findall('Child',str(x))))
df['n_victim_Teen']=df['victim_age_group'].map(lambda x:len(re.findall('Teen',str(x))))
df['n_victim_Adult']=df['victim_age_group'].map(lambda x:len(re.findall('Adult',str(x))))
```

```python
status=df['participant_status'].map(lambda x:re.findall('[a-zA-Z]+',str(x)))
len_each=status.map(lambda x:len(x))
idx=len_each[len_each==0].index.values
status[idx]=df.iloc[idx]['participant_status']
vic_status={}
suspect_status={}
for i in range(len(status)):
    j=int(df['n_suspect'][i])
    k=int(df['n_victim'][i])
    if i not in idx:
        if j==0:
            suspect_status[i]=NaN
        else:
            suspect_status[i]=status[i][-j:]
        if k==0:
            vic_status[i]=NaN
        else:
            vic_status[i]=status[i][0:k]
suspect_status=Series(suspect_status)
vic_status=Series(vic_status)
df['victim_status']=vic_status
df['suspect_status']=suspect_status
```

```python
df['n_victim_Killed']=df['victim_status'].map(lambda x:len(re.findall('Killed',str(x))))
df['n_victim_Injured']=df['victim_status'].map(lambda x:len(re.findall('Injured',str(x))))
df['n_victim_Unharmed']=df['victim_status'].map(lambda x:len(re.findall('Unharmed',str(x))))
df['n_suspect_Killed']=df['suspect_status'].map(lambda x:len(re.findall('Killed',str(x))))
df['n_suspect_Injured']=df['suspect_status'].map(lambda x:len(re.findall('Injured',str(x))))
df['n_suspect_Unharmed']=df['suspect_status'].map(lambda x:len(re.findall('Unharmed',str(x))))
df['n_suspect_Arrested']=df['suspect_status'].map(lambda x:len(re.findall('Arrested',str(x))))
```

```python
mass=df['incident_characteristics'].map(lambda x:re.findall('Mass Shooting',str(x)))
len_each=mass.map(lambda x:len(x))
idx=len_each[len_each==0].index.values
df['mass_shooting']=1
df['mass_shooting'][idx]=0
rob=df['incident_characteristics'].map(lambda x:re.findall('robbery',str(x)))
len_each=rob.map(lambda x:len(x))
idx=len_each[len_each==0].index.values
df['robbery']=1
df['robbery'][idx]=0
drive=df['incident_characteristics'].map(lambda x:re.findall('Drive-by',str(x)))
len_each=drive.map(lambda x:len(x))
idx=len_each[len_each==0].index.values
df['Drive-by']=1
df['Drive-by'][idx]=0
df.head()
```

```python
date = pd.to_datetime(df['date'])
df["quarter"] = date.map(lambda x:x.quarter)
df["month"] =date.map(lambda x:x.month)
df["dayofweek"] =date.map(lambda x:x.dayofweek)
```
